```python
#1
def selection_sort1(my_list):
  for i in range(len(my_list)-1):
    for j in range(len(my_list)-i-1):
      if my_list[j]>my_list[j+1]:
        temp=my_list[j]
        my_list[j]=my_list[j+1]
        my_list[j+1]=temp
  return my_list
print(selection_sort1(my_list = [10, 1, 20, 3, 6, 2, 5, 11, 15, 2, 12, 14, 17, 18, 29]))       
```

```python
#2
def selection_sort2(my_list):
  for i in range (0,len(my_list)-1):
    min=my_list[i]
    min_i=i
    for j in range(i+1,len(my_list)):
      if my_list[j]<min:
        min=my_list[j]
        min_i=j
    temp=min
    my_list[min_i]=my_list[i]
    my_list[i]=temp
  return my_list
print(selection_sort2(my_list =[10, 1, 20, 3, 6, 2, 5, 11, 15, 2, 12, 14, 17, 18, 29]))               
```

```python
#9
import math

print('Enter points')
n = int(input())

lst2 = []   # to store the distances
lst = []    # to store the entered points

for i in range(n):
    x , y = map(int,input().split())
    temp = (x,y)
    dist = math.sqrt(x*x + y*y)
    lst2.append(dist)
    lst.append(temp)

print('Entered points = ',lst)
print('\nMinimum distance = ', min(lst2))
print('\nHere the closest point is', lst[lst2.index(min(lst2))],'which has a distance of', min(lst2),'from the origin.')
```
