```python
#2
def fib(num):
    first=0
    sec=1
    print(first,sec,end=" ")
    fib=first+sec
    val = 0 
    while fib<=num:
      val=first+sec
      print(val,end=" ")
      first=sec
      sec=val
      fib=first+sec
fib(10)       
```

```python
#4
def sf(str1):
  up_count=0
  low_count=0
  for i in str1:
    ascii=ord(i)
    if 97<=ascii<=122:
      low_count+=1
    elif 65<=ascii<=90:
      up_count+=1
  print("No. of Uppercase characters:",up_count)
  print("No. of Lowercase Characters:",low_count)
sf("The quick Sand Man")  
```

```python
#7
def show_palindrome(num):
    for i in range(1, num):
        print(i, end = "")
    for rev_i in range(num, 0, -1):
        print(rev_i, end = "")
    print("")
show_palindrome(3)
```

```python
#8
def show_palindromic_triangle(n):
    for row in range(1, n+1):
        for i in range(1, (n-row)+1):
            print(" ",end = " ")
        for a in range(1, row):
            print(a, end = " ")
        for b in range(row, 0, -1):
            print(b, end = " ")
        print("") 
show_palindromic_triangle(5)   
```

```python
#9
import math

def area_circumference_generator(radius):
    circumference = 2 * math.pi * radius
    area = math.pi * radius**2
    area_circumference = (area, circumference)
    return area_circumference

area, circumference = area_circumference_generator(1)
print("area is",area,"and circumference is",circumference)
```

```python
#10
def make_sq(var):
  out={}
  start,end=var
  for i in range(start,end+1):
    valu=i**2
    out[i]=valu
  return out
make_sq((1,3))    
```

```python
#11
def rem_duplicate(tup):
    tup_list = list(tup)
    no_duplicate = []
    for item in tup_list:
        if(item not in no_duplicate):
            no_duplicate.append(item)
    fresh_tup = tuple(no_duplicate)
    return fresh_tup
rem_duplicate((1,1,1,2,3,4,5,6,6,6,6,4,0,0,0))
```

```python
#13
def basic_calc(operator, num_1, num_2):
    if(operator == "+"):
        return num_1 + num_2
    
    elif(operator == "-"):
        return num_1 - num_2
    
    elif(operator == "/"):
        return num_1 / num_2
    
    elif(operator == "*"):
        return num_1 * num_2
basic_calc("+", 10, 20)      
```

```python
#14
def char_remover(sentence, position):
    new_sentence = sentence[0]
    removed_part = ""
    
    for index in range(1, len(sentence)):
        if(index % position != 0):
            new_sentence += sentence[index]
        else:
            removed_part += sentence[index]
    new_sentence += removed_part
            
    return new_sentence
char_remover("I love programming.", 3)
```

```python
#15
def food_panda(food, destination = "Dhanmondi"):
    price = {"Rice": 105, "Potato": 20, "Chicken": 250, "Beef": 510, "Oil": 85}
    total = 0
    for item in food:
        total += price[item]
    if(destination == "Dhanmondi"):
        total += 30
    else:
        total += 70
    return total
food_panda(["Rice", "Beef", "Rice"])
```
