```python
#11
s= input()
s1=s.replace('"',"")
s2=s1.replace(" ","")
s3=s2.lower()
dict = {}

for char in s3:
    if(char in dict):
        dict[char] += 1
        continue
    dict[char] = 1
    
    
print(dict)
```
